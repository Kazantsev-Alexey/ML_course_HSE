## Задание 2 (1 балл)

Найдите экстремумы этих функций в Python.

```python
import numpy as np
import sympy as sp

x = sp.Symbol('x', real=True)

def critical_points_solver(x,y):
    derivative = sp.diff(y, x)
    second_derivative = sp.diff(derivative, x)
    critical_points = sp.solve(sp.Eq(derivative, 0), x)
    for point in critical_points:
        second_derivative_value = second_derivative.subs(x, point)
        if second_derivative_value > 0:
            ext_type = "минимум в точке:"
        elif second_derivative_value < 0:
            ext_type = "максимум в точке:"
        print(ext_type, f'x = {point}')

print('a)')
critical_points_solver(x, (x**2 - 30*x + 1))
print('b)')
critical_points_solver(x, (-x**3 + 5*x - 7*x**2 + 17))
print('c)')
critical_points_solver(x, (x**2 - 1/x))
print('d)')
critical_points_solver(x, (1/x**5 + x**17 + 8))
```

## Задание 4 (1 балла)

Найдите точки перегиба этих функций в Python.

```python
def inflection_points_solver(x,y):
    derivative = sp.diff(y, x)
    second_derivative = sp.diff(derivative, x)
    inflection_points = sp.solve(sp.Eq(second_derivative, 0), x)
    if len(inflection_points) > 0:
        inflection_points = [float(x) for x in inflection_points]
        return inflection_points
    else:
        return None
a = inflection_points_solver(x, (x**2 - 30*x + 1))
b = inflection_points_solver(x, (-x**3 + 5*x - 7*x**2 + 17))
c = inflection_points_solver(x, (x**2 - 1/x))
d = inflection_points_solver(x, (1/x**5 + x**17 + 8))
print('a',a)
print('b',b)
print('c',c)
print('d',d)
```

```python
import matplotlib.pyplot as plt
from sympy import symbols, diff, solve

def solver_and_graph(x, y,x_range):
    # x_range=[-10,10]
    functions = lambda t: float(y.subs(x, t))

    derivative = sp.diff(y, x)
    second_derivative = sp.diff(derivative, x)
    x_critical_points = [float(x) for x in sp.solve(sp.Eq(derivative, 0), x)]
    y_critical_points = [functions(z) for z in x_critical_points]
    x_inflection_points = [float(x) for x in sp.solve(sp.Eq(second_derivative, 0), x)]
    y_inflection_points = [functions(z) for z in x_inflection_points]



    x_values = np.linspace(x_range[0], x_range[1], 100)
    functions = lambda t: float(y.subs(x, t))
    y_values = [functions(z) for z in x_values]
    plt.figure(figsize=(8, 6))
    plt.plot(x_values, y_values, label='f(x)')
    plt.scatter(x_critical_points, y_critical_points, color='red', label='Critical points', )

    # Добавляем точки перегиба
    plt.scatter(x_inflection_points, y_inflection_points, color='green', label='Inflection points', )

    # для подписей делаем смещение в масштабе самих значений + смещение на 10
    for i, (xc, yc) in enumerate(zip(x_critical_points, y_critical_points)):
        plt.annotate(f"({xc:.2f}, {yc:.2f})", xy=(xc, yc), xytext=(abs(xc * 0.5)+10, abs(yc * 0.3)+10),
                     textcoords='offset points', arrowprops=dict(arrowstyle='->', lw=0.5), fontsize=9)

    for i, (xi, yi) in enumerate(zip(x_inflection_points, y_inflection_points)):
        plt.annotate(f"({xi:.2f}, {yi:.2f})", xy=(xi, yi), xytext=(xi*1.3+2, yi*1.3+2),
                     textcoords='offset points', arrowprops=dict(arrowstyle='->', lw=0.5), fontsize=9)

    plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
    plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
    plt.title(f"График функции {y} с экстремумами и точками перегиба")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend()
    plt.grid(True)
    plt.show()

solver_and_graph(x, (x**2 - 30*x + 1),[-20,50])
solver_and_graph(x, (-x**3 + 5*x - 7*x**2 + 17),[-10,10])
solver_and_graph(x, (x**2 - 1/x),[-5,5])
solver_and_graph(x, (1/x**5 + x**17 + 8),[-3,3])
```

## Задание 6 (1 балл)

Модифицируйте функцию, осуществляющую метод градиентного спуска с семинара: добавьте опцию останавливать метод, если абсолютное значение производной функции в точке < некоторого $\delta$ (delta добавьте в аргументы функции).

```python
# запишем функции, чтобы вызывать их коротко
def f_a(x):
  return (x**2 - 30*x + 1)
def f_b(x):
  return -x**3 + 5*x - 7*x**2 + 17
def f_c(x):
  return x**2 - 1/x
def f_d(x):
  return (1/x**5 + x**17 + 8)


def gradient_descent(x_start, learning_rate, epsilon, num_iterations, f, delta=0.01):
    x_curr = x_start
    derivative = sp.diff(f(x))

    trace = []
    trace.append(x_curr)

    for i in range(num_iterations):
        #посчитаем градиент отдельно, чтобы добавить условия для дельты и ограничение сверху для работы с Simpy
        grad = max(min(derivative.subs(x, x_curr), 1e6), -1e6)
        x_new = x_curr - learning_rate * grad
        trace.append(x_new)
        if abs(x_new - x_curr) < epsilon or abs(grad) < delta:
            return x_curr, trace


        x_curr = x_new

    return x_curr, trace

def graph(function, trace, xmin, x_linspace=[-10,10,100]):
    x_values = [x for x in np.linspace(x_linspace[0], x_linspace[1], x_linspace[2])]
    f_values = [function(x) for x in x_values]

    plt.figure(figsize=(6,6))
    trace_values = [function(x) if x<x_linspace[1] and x >x_linspace[0] else None for x in trace ]

    # для подписей делаем смещение в масштабе самих значений + смещение на 10
    plt.annotate(f"({xmin:.2f}, {function(xmin):.2f})", xy=(xmin, function(xmin)), xytext=(abs(xmin * 0.5)+10, abs(function(xmin) * 0.3)+10),
                     textcoords='offset points', arrowprops=dict(arrowstyle='->', lw=0.5), fontsize=9)

    plt.scatter(trace, trace_values, c='red', label='Trace points')
    plt.scatter(xmin, function(xmin), c='green', label='Min point',)

    plt.axvline(x=0, c = 'black')
    plt.axhline(y=0, c = 'black')
    plt.plot(x_values, f_values)
    plt.title(f'График функции ')

    plt.legend()
    plt.xlabel('x')
    plt.ylabel('f(x)')

    plt.show()

# выполняем gd
xmin_a, trace_a = gradient_descent(-75, 0.051, 0.001, 100, f_a)
xmin_b, trace_b = gradient_descent(-100, 0.001, 0.001, 100, f_b)
xmin_c, trace_c = gradient_descent(100, 0.03, 0.001, 100, f_c)
xmin_d, trace_d = gradient_descent(2, 0.000001, 0.001, 100, f_d)

# отрисовываем
graph(f_a,trace_a,xmin_a,[-100,100,1000])
graph(f_b,trace_b,xmin_b)
graph(f_c,trace_c,xmin_c)
graph(f_d,trace_d,xmin_d,[-5,5,1000])
```
